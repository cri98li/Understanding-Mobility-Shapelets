```python
elabora = "elabora"
```

```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir(elabora) if isfile(join(elabora, f))]
```

```python
import pandas as pd
```

```python
def converti(file, colonne):
    df = pd.read_csv(file, header=None)
    df[colonne].to_csv(file, header=None, index=True)
    
for f in onlyfiles:
    print(f)
    converti(join(elabora,f), [0])
```

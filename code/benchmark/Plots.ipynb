```python
import pandas as pd
import numpy as np
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
import os
from glob import glob
from matplotlib.lines import Line2D
```

```python
glob("../../data/*")
```

```python
df = pd.read_csv("../../data/simple_shape.zip")

df
```

```python
def plot(df, tid, color, alpha, offset_x=0, offset_y=0):
    df_plot = df[df.tid == tid]
    plt.plot(df_plot.lat - df_plot.lat.iloc[0] + offset_x, df_plot.lon - df_plot.lon.iloc[0] + offset_y, 
             label=df_plot.y.iloc[0], c=color, alpha=alpha)
    
alpha = .05
for i in range(1, 10+1):
    plot(df, i, "tab:orange", alpha=alpha*i, offset_x=5)
    
for i in range(1, 10+1):
    plot(df, i+100, "tab:green", alpha=alpha*i, offset_x=2)

plt.xlim(-0.5, 10.5)
plt.ylim(-0.5, 10.5)
plt.xticks([])
plt.yticks([])
plt.gca().set_box_aspect(1)
custom_lines = [Line2D([0], [0], color="tab:orange"),
                Line2D([0], [0], color="tab:green")]
plt.legend(custom_lines, ["right", "straight"], fontsize=13, frameon=False, loc='lower right', bbox_to_anchor=(1.04, -0.03))
plt.title("SR Dataset", fontsize=16)
plt.savefig("Example_simple_shape.png",  bbox_inches='tight')
```

```python
alpha = .5
for i in range(0, 10+1):
    plot(df, i, "tab:orange", alpha=alpha, offset_x=5)
    
for i in range(1, 10+1):
    plot(df, i+100, "tab:green", alpha=alpha, offset_x=2)
    
for i in range(1, 10+1):
    plot(df, i+200, "tab:blue", alpha=alpha, offset_x=2)

#plt.xlim(-0.5, 10.5)
#plt.ylim(-0.5, 10.5)
plt.xticks([])
plt.yticks([])
plt.gca().set_box_aspect(1)
custom_lines = [Line2D([0], [0], color="tab:orange"),
                Line2D([0], [0], color="tab:green"),
                Line2D([0], [0], color="tab:blue")]
plt.legend(custom_lines, ["straight", "circling","bending"], fontsize=13, frameon=False, loc='lower right', bbox_to_anchor=(1.04, -.03))
plt.title("Yao et al. Dataset", fontsize=16)
plt.savefig("Example_yao.png",  bbox_inches='tight')
```

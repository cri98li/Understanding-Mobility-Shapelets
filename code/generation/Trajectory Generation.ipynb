```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def generate_trajectory(n_steps, smoothness=11, turning_angle=np.pi/2, step_size=0.1, noise_weight=.5):
    x, y = (random.random() -.5)*noise_weight, (random.random() -.5)*noise_weight
    trajectory = [(x, y)]
    
    angle=np.pi/2
    smoothness_steps = 0
    
    for i in range(n_steps-1):
        if i*2 > n_steps-smoothness and smoothness_steps < smoothness:
            angle -= turning_angle / smoothness
            smoothness_steps += 1
        
        x += step_size * np.cos(angle)
        y += step_size * np.sin(angle)
        trajectory.append((x, y))
    
    trajectory = np.array(trajectory)
    trajectory += (np.random.rand(*trajectory.shape) - .5) * noise_weight * step_size
    
    return trajectory
```

```python
for i in range(10):
    trajectory = generate_trajectory(50, smoothness=20, turning_angle=-np.pi/2)
    plt.plot(trajectory[:, 0], trajectory[:, 1])
    plt.scatter(trajectory[:1, 0], trajectory[:1, 1])
```
